```python
import pandas as pd
import numpy as np

file_path = '../data/Churn_Modelling.csv'
df = pd.read_csv(file_path)

print("Dataframe shape: \n", df.shape)
print("\nFirst 5 rows: \n", df.head())
print("\nLast 5 rows: \n", df.tail())
print("\nData types and non-null counts: \n", df.info())
print("\nSummary Statistics for numerical columns: \n", df.describe())
print("\nNumber of missing values per column: \n", df.isnull().sum())
print("\nNumber of duplicated rows: \n", df.duplicated().sum())
```

```python
# Data cleaning
df_cleaned = df.drop(['RowNumber', 'Surname'], axis=1)
# no missing or duplicated data
```

```python
# exploratory analysis
print('\nUnique values in Geography: \n', df_cleaned['Geography'].unique())
print('\nNumber of values in Geography: \n', df_cleaned['Geography'].value_counts())
print('\nNumber of values in Gender: \n', df_cleaned['Gender'].value_counts())
print('\nNumber of values in Exited: \n', df_cleaned['Exited'].value_counts())
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+mysqlconnector://root:123456@localhost:3306/bca')
df_cleaned.to_sql('customers', con=engine, if_exists='append', index=False)
print(f"\nSuccessfully loaded {len(df_cleaned)} rows into 'customers' table using SQLAlchemy.")
```
